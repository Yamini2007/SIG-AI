```python
#decision tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head
```

```python
# Features and target
X = df.drop('target', axis=1)
y = df['target']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
#used to train model to predict labels for test feauters(X_test)
#these predictions are stored in y_pred

print("Accuracy:", accuracy_score(y_test, y_pred))
#compares predicted labels(y_pred)with actual labels(y_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
#shows matrix comparingf true labels(y_test) and predicted labels(y_pred)
```

```python
#SVM --Support vector Machine
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

```

```python
#initializing and training svm model
svm_model = SVC(kernel='linear')  # You can also try 'rbf', 'poly', etc.
svm_model.fit(X_train, y_train)
```

```python
y_pred_svm = svm_model.predict(X_test)

print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("\nSVM Classification Report:\n", classification_report(y_test, y_pred_svm))
print("\nSVM Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
#i.e 95.6%accuracy on test data
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
#initializing and training the model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
#prdeict on test data
rf_pred = rf_model.predict(X_test)
#stores predicted data in rf_pred
```

```python
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print("\nRandom Forest Classification Report:\n", classification_report(y_test, rf_pred))
print("\nRandom Forest Confusion Matrix:\n", confusion_matrix(y_test, rf_pred))
```

```python
#starting hyperparameter tuning for Decision Tree using GridSearchCV 
#Hyperparameter tuning helps you:Improve accuracy;Avoid overfitting or underfitting;Choose the best settings for the algorithmfrom sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
#defining Parameter Grid
param_grid = {
    'max_depth': [2, 4, 6, 8, 10, None],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}
```

```python
#initialising GridSearchCV
grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,               # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1           # Use all processors
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Tuned Decision Tree Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
