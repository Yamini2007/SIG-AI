```python
#aim:Predict line that best fits our data(y=m*x+b)
import numpy as np
#numpy--mathematical operations
#matplotlib.pyplot--to plot graph from matplot library and model visually
import matplotlib.pyplot as plt
```

```python
# Creating  sample data (X and y)
X = np.array([1, 2, 3, 4, 5])   # Features (input)
y = np.array([2, 4, 6, 8, 10])  # Labels (output)
# modeling y=2*x linear relationship
```

```python
# Initializing parameters
m = 0  # slope
b = 0  # intercept
# i.e intial guess
learning_rate = 0.01
#learning_rate--conntrols how big each update step  is
epochs = 1000  # Number of times we updating m and b
n = len(X)
```

```python
#  Performing Gradient Descent
for i in range(epochs):
    y_pred = m * X + b  # Prediction using current m and b
    error = y - y_pred  # Difference between actual and predicted values

    # Gradients for m and b
    dm = (-2 / n) * sum(X * error)
    db = (-2 / n) * sum(error)

    # Update m and b
    m -= learning_rate * dm
    b -= learning_rate * db
# for every epoch we need to calculate the gradient and update m and b
```

```python
# Printing final model
print(f"Linear Regression Model: y = {m:.2f}x + {b:.2f}")
```

```python
# Ploting the result
plt.scatter(X, y, label="Actual Data")  # Dots
plt.plot(X, m * X + b, color='red', label="Predicted Line")  # Line
plt.legend()
plt.title("Linear Regression from Scratch")
plt.show()
```

```python
# K-Means clustering(unspervised Learning)
#Aim--Group data into clusters 
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(n_samples=100, centers=2, n_features=2, random_state=42)
#make_blobs -- It's a function that creates clusters known as blobs of points for testing ML models
#n_sampples=100 --Generate 100 total data points
#centers=2 --Create 2 natural clusters
#n_feauters=2 -- Each point have 2 feauters(x and y)
#random_state=42 -- i.e gives ame result every time
#why 42?? --its a random Numbeer;but its a fun tradition in Programming world
```

```python
# Initializing number of clusters (k) and picking random centroids
k = 2
centroids = X[np.random.choice(range(len(X)), k, replace=False)]
#X is a dataset (i.e 100 points in 2D)
#np.random.choice(...)--Randomly picks k  indices from the data
#replace=False-- Ensuring that no repeated points i.e each choosen point is unique
```

```python
k = 2

# Step 1: Initialize centroids randomly
centroids = X[np.random.choice(len(X), k, replace=False)]

# Step 2: Run K-Means manually
for _ in range(10):
    clusters = []

    # Assign each point to the nearest centroid (manual way)
    for point in X:
        distances = []
        for centroid in centroids:
            # Euclidean distance manually
            dist = np.sqrt((point[0] - centroid[0])**2 + (point[1] - centroid[1])**2)
            distances.append(dist)
        cluster = np.argmin(distances)  # Gives the Index of closest centroid
        #arg min is used to give the index of smallest value in list or array
        clusters.append(cluster)

    clusters = np.array(clusters)

    # Step 3: Update centroids
    new_centroids = []
    for i in range(k):
        # Take all points in cluster i
        points_in_cluster = X[clusters == i]
        # Compute the mean
        if len(points_in_cluster) > 0:
            new_centroid = points_in_cluster.mean(axis=0)
        else:
            new_centroid = centroids[i]  # No change if empty cluster
        new_centroids.append(new_centroid)

    new_centroids = np.array(new_centroids)

    # Stop if centroids didn't change
    if np.all(centroids == new_centroids):
        break

    centroids = new_centroids

#After finishing the loop, each point is assigned to a cluster.;Final centroids and cluster assignments are obtained.

```

```python
# Ploting the clusters
for i in range(k):
    plt.scatter(X[clusters == i][:, 0], X[clusters == i][:, 1], label=f"Cluster {i+1}")
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, color='black', label='Centroids')
plt.legend()
plt.title("K-Means Clustering from Scratch")
plt.show()
```
